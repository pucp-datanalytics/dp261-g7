# Proyecto 10: Retención de Talento Empresarial (HR Attrition)

## 1. Contexto de la Industria
La retención de talento se ha convertido en un desafío crítico en la gestión de Recursos Humanos. Se observa que el costo de reemplazo de un empleado especializado puede oscilar entre el **50% y el 200%** de su salario anual, debido a los costos asociados a procesos de selección, incorporación (onboarding) y la curva de aprendizaje inicial. La reducción de la tasa de rotación impacta directamente en los márgenes de rentabilidad y en el *know-how* estratégico de la organización.

## 2. Definición del Problema de Negocio
Una empresa multinacional enfrenta actualmente una **tasa de rotación (turnover) del 16%**. El departamento de Recursos Humanos necesita implementar un mecanismo proactivo en lugar de reactivo. 
**Problema:** ¿Cómo identificar qué empleados tienen alta probabilidad de abandonar la empresa de manera temprana, antes de que formalicen su renuncia?

Al responder esto, la compañía podrá focalizar planes de retención, ajustes salariales, programas de mentoría o mejoras en el clima laboral de forma quirúrgica sobre el talento de alto riesgo.

## 3. Variable Objetivo (Target)
El problema se aborda bajo un esquema de **Clasificación Binaria**.
* **Variable Dependiente (`attrition`):** 
   - `1`: El empleado abandonó la compañía.
   - `0`: El empleado permanece activo en la compañía.

## 4. KPIs y Criterios de Éxito del Modelo

### KPIs de Negocio (Business Value)
1. **Disminución de Rotación:** Reducir la tasa global de abandono del 16% a al menos un 12% en un periodo de 12 meses.
2. **Reducción de Costos Operativos:** Lograr un 20% de ahorro en los presupuestos anuales de selección (incluyendo métricas asociadas como *Time-to-Fill*).

### Criterios de Éxito Matemáticos (Modelo Analítico)
1. **Métrica Principal:** Priorizaremos rigurosamente el **Recall (Sensibilidad)** para la clase `1` (Fuga), por encima de la Precisión. El costo de los Falsos Negativos (no detectar a un empleado que sí va a renunciar y perderlo) es el riesgo principal del negocio. Evaluaremos que el modelo consiga un `Recall > 80%` para considerar el Sprint 4 y 5 como exitoso.
2. **Métrica de Balance:** Observaremos el **F1-Score** y el cruce **AUC-ROC ( > 0.75)** para evaluar la separación confiable de clases y evitar un disparo innecesario de falsos positivos masivos.

## 5. Requerimientos de los Stakeholders
* **Director de RRHH y Managers Internos:** Necesitan *interpretabilidad* del modelo. Saber **por qué** un empleado está en riesgo de fugarse es tan importante como la alerta misma (por ejemplo: poder diferenciar si el riesgo algorítmico fue subido artificialmente por su nivel de salario, bajo involucramiento en proyectos o bajo nivel de satisfacción).

## 6. Glosario de Términos
* **Attrition (Rotación natural/fuga):** Reducción de la plantilla por causas como renuncias, jubilaciones, etc. En el contexto de este proyecto, nos enfocamos en bajas voluntarias prevenibles.
* **Turnover Rate:** Mide la porción de empleados que se van respecto del promedio total de empleados de la multinacional.
* **Retention Strategies:** Mecanismos accionables (entrenamiento, monetarios o emocionales) aplicados por la firma hacia el cluster de riesgo.

## 7. Referencias Bibliográficas y del Dataset
- Fuente Principal: Dataset Comercial de Recursos Humanos.
- Dimensiones del estudio: Base de 14,000 registros, 20 variables de características demográficas, laborales y encuestas de satisfacción. Consta de 8 departamentos analizados.
- Rúbrica metodológica adaptada: "Cartera de Proyectos PUCP - S09", Proyecto 10.
